## Purpose of script:
#### Light weight market basket analysis for dockerization

```python
import numpy as np
import pandas as pd
from apyori import apriori
import pickle
```

```python
orig = pd.read_csv('../Datasets/Market_Basket_Optimisation.csv', header=None)
print(orig.shape)
orig.head(2)
```

```python
orig.info()
```

```python
list_of_lists = []

for i in range(orig.shape[0]):
    basket = [str(orig.values[i, j]) for j in range(orig.shape[1])]
    basket = [i for i in basket if i != 'nan']
    list_of_lists.append(basket)
    
print(list_of_lists[0])
print('\n')
print(list_of_lists[-1])
```

```python
rules = apriori(list_of_lists, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
rules = list(rules)

print(len(rules))
for i in rules[0]:
    print(i)
```

```python
Antecedents = []
Consequents = []
Rule = []
Support = []
Confidence = []
Lift = []
for item in rules:
    pair = item[0]
    items = [x for x in pair]
    Antecedents.append(item[2][0][0])
    Consequents.append(item[2][0][1])
    Rule.append(items)
    Support.append(str(item[1]))
    Confidence.append(str(item[2][0][2]))
    Lift.append(str(item[2][0][3]))
```

```python
rule_dict = {'antecedents': Antecedents, 'consequents': Consequents, 
             'rule': Rule, 'support': Support, 'confidence': Confidence, "lift": Lift} 

df_assoc = pd.DataFrame(rule_dict)
df_assoc.sort_values(by='support', ascending=False, inplace=True)
df_assoc.head(20)
```
