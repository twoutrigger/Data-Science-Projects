```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

## Tutorial example

```python
# Set the shape paremeters
#
a, b = 90, 10
#
# Generate the value between
#
x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
#
# Plot the beta distribution
#
plt.figure(figsize=(7,7))
plt.xlim(0.7, 1)
plt.plot(x, beta.pdf(x, a, b), 'r-')
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

## My comparison example

```python
## My example to compare two beta distributions

# Set the shape paremeters
a_1, b_1 = 6, 4
a_2, b_2 = 30, 20

# Generate the value between
x_1 = np.linspace(beta.ppf(0.01, a_1, b_1),beta.ppf(0.99, a_1, b_1), 100)
x_2 = np.linspace(beta.ppf(0.01, a_2, b_2),beta.ppf(0.99, a_2, b_2), 100)

# Plot the beta distribution
plt.figure(figsize=(7,7))
plt.xlim(0, 1)
plt.plot(x_1, beta.pdf(x_1, a_1, b_1), 'r-')
plt.plot(x_2, beta.pdf(x_2, a_2, b_2), 'g-')
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability Density', fontsize='15')
plt.show()
```
