```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Crude Soybean Oil Futures May 2019 dataset 

dorig = pd.read_csv('soybean_052019.csv',
                    header=0, 
                    index_col='Date', 
                    parse_dates=True)
dorig.head()
```

```python
dorig.describe()
```

```python
dorig['Close'].plot(grid=True)
```

```python
dorig[['Open','Close']].plot(grid=True)
```

```python
dorig[['High','Low']].plot(grid=True)
```

```python
dorig[['Volume','Open Interest']].plot(grid=True)
```

```python
sns.factorplot(x='Close',y='Settle',data=dorig,kind='strip', aspect=2)
```

```python
sns.factorplot(x='Open',y='Pre Settle',data=dorig,kind='strip',aspect=2)
```

```python
sns.lmplot(x='Close',y='Settle', data=dorig, aspect=2)
```
