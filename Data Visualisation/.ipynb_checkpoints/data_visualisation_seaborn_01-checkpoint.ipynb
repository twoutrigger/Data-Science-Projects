```python
## script uses toy dataset (Abalone) to document some useful Seaborn data visualisation tools
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
## column names and explanation from abalonenames.txt
cols = ['Sex', 'Length', 'Diameter', 'Height', 'Weight', 'Shucked',
        'Viscera', 'Shell', 'Rings']

orig = pd.read_csv('../Datasets/abalonedata.csv', names=cols, header=None)

print(orig.shape)

orig.head()
```

```python
orig.describe()
```

```python
# Per documentation "I" is infant
orig.Sex.value_counts()
```

## Joint Plot

```python
sns.jointplot(x="Length", y="Rings", data=orig, size=5)
```

## Scatter Plot

```python
sns.FacetGrid(orig, hue="Sex", height=5) \
   .map(plt.scatter, "Length", "Rings") \
   .add_legend()
```

## Box Plot

```python
## visible outliers in the Height variable
sns.boxplot(x="Sex", y="Height", data=orig)
```

```python
ax = sns.boxplot(x="Sex", y="Height", data=orig)
ax = sns.stripplot(x="Sex", y="Height", data=orig, jitter=True, edgecolor="gray")
```

## Violin Plot

```python
sns.violinplot(x="Sex", y="Height", data=orig, size=6)
```

## KDE Plot

```python
sns.FacetGrid(orig, hue="Sex", height=6) \
   .map(sns.kdeplot, "Diameter") \
   .add_legend()
```

## Pair Plot

```python
sns.pairplot(orig)
```

```python
sns.pairplot(orig, hue='Sex')
```
