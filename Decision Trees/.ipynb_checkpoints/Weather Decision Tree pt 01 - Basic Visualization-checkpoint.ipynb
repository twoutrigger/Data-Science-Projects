```python
import sas7bdat
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

```python
bat_file = sas7bdat.SAS7BDAT('weather.sas7bdat')

worig = bat_file.to_data_frame()

print(f'Dataset shape is: {worig.shape}')

worig.head()
```

```python
for i in list(worig.columns):
    print(worig[i].value_counts())
```

```python
def target_binary(x):
    if (x == 'yes') or (x == 't'):
        return 1
    elif (x == 'no') or (x == 'f'):
        return 0
```

```python
wedit = worig.copy()

wedit = pd.get_dummies(wedit,columns=['Outlook','Temp','Humidity'])
wedit['Windy'] = wedit['Windy'].apply(target_binary)
wedit['Play'] = wedit['Play'].apply(target_binary)
wtar = wedit.pop('Play')

wedit.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(wedit,wtar,
                                                    test_size=0.25,
                                                    random_state=100)

print(f'X_train: {X_train.shape} X_test: {X_test.shape} Y_train: {Y_train.shape} Y_test: {Y_test.shape}')
```

```python
w_entr = DecisionTreeClassifier(criterion='entropy', max_depth=8, min_samples_leaf=1)
w_entr.fit(X_train, Y_train)
```

```python
dot_data = StringIO()

export_graphviz(w_entr, out_file=dot_data, feature_names=wedit.columns, 
                filled=True, rounded=True,
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
w_gini = DecisionTreeClassifier(criterion='gini', max_depth=8, min_samples_leaf=1)
w_gini.fit(X_train, Y_train)
```

```python
dot_data = StringIO()

export_graphviz(w_gini, out_file=dot_data, feature_names=wedit.columns, 
                filled=True, rounded=True,
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
pred_entr = w_entr.predict(X_test)
pred_gini = w_gini.predict(X_test)

print(f'Entropy accuracy is {accuracy_score(Y_test,pred_entr)*100}%')
print(f'Gini accuracy is {accuracy_score(Y_test,pred_gini)*100}%')
```
