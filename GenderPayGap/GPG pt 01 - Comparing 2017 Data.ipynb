```python
# Median Hourly distribution
# checking if the tails made it into website
plt.figure(figsize=(20, 10))
plt.hist(porig['DiffMedianHourlyPercent'], bins=300)
plt.show()
```

```python
# upper cutoff is MALLING HEALTH (UK) LIMITED at 73%
porig.iloc[porig['DiffMedianHourlyPercent'].nlargest(5).index,:]
```

```python
# lower cutoff is INBRELLA LIMITED capped at 100%
porig.iloc[porig['DiffMedianHourlyPercent'].nsmallest(5).index,:]
```

```python
# perfect zero accounts for 8% of distribution
len(porig[porig['DiffMedianHourlyPercent'] == 0])/porig.shape[0]
```

```python
plt.figure(figsize=(20, 10))
plt.boxplot(porig['DiffMedianHourlyPercent'])
plt.show()
```

```python
pedit = porig.copy()
pedit = pedit[['EmployerName', 'CompanyNumber', 'DiffMedianHourlyPercent']]

pedit.head()
```

```python
# some outliers removed
pedit = pedit[(pedit['DiffMedianHourlyPercent'] <= 73.0) & (pedit['DiffMedianHourlyPercent'] >= -164.0)]

print(pedit.shape)
```

```python
# identifying repeats
pedit['EmployerName'].value_counts()[:5]
```

```python
pedit[pedit['EmployerName'] == 'Talbot Underwriting Services Ltd']
```

```python
# confirm there is no significant data difference
porig[porig['EmployerName'] == 'Talbot Underwriting Services Ltd']
```

```python
pedit.drop(8745, inplace=True)

pedit.iloc[8740:8750, :]
```

```python
pedit.reset_index(drop=True, inplace=True)

pedit.tail(1)
```

```python
pedit['DiffMedianHourlyPercent'].nsmallest(3)
```

```python
pedit.iloc[pedit['DiffMedianHourlyPercent'].nsmallest(3).index,2] = -100.0

pedit.iloc[pedit['DiffMedianHourlyPercent'].nsmallest(5).index,:]
```

```python
pedit['CompanyNumber'].fillna(value=0, inplace=True)

pedit.info()
```

```python
print(f'Orig lowercase: {pedit.iloc[10548, 0]}')
pedit['EmployerName'] = pedit['EmployerName'].str.upper()
print(f'Post edit: {pedit.iloc[10548, 0]}')
```

```python
pedit.sort_values(by=['DiffMedianHourlyPercent'], ascending=True, inplace=True)

pedit.head(5)
```
