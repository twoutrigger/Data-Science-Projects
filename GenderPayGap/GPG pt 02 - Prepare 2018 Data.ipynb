## Compare 2017 and 2018

```python
porig = pd.read_csv('UK Gender Pay Gap Data - 2017 to 2018.csv', header=0)

print(porig.shape)
porig.head(2)
```

```python
# verify that columns for 2017 & 2018 are the same
blank_list =[]

for i in range(len(norig.columns)):
    if norig.columns[i] != porig.columns[i]:
        blank_list.append[i]
        
print(blank_list)
```
