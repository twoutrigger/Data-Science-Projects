```python
aorig = pd.read_csv('2017clean1.csv', header=0, index_col=0)
borig = pd.read_csv('2018clean1.csv', header=0, index_col=0)

print(aorig.shape)
print(borig.shape)
aorig.head(2)
```

```python
def postcode(x):
    x = x.upper()
    found = re.findall(r'\b[A-Z]{1,2}[0-9][A-Z0-9]? [0-9][ABD-HJLNP-UW-Z]{2}\b', x)
    output = None
    
    if len(found) > 0:
        output = found[0]
    else:
        found = re.findall(r'\b[A-Z]{1,2}[0-9][A-Z0-9]?[0-9][ABD-HJLNP-UW-Z]{2}\b', x)
        if len(found) > 0:
            found = str(found[0])
            if len(found) % 2 == 0:
                output = found[:int(len(found)/2)] + ' ' + found[int(len(found)/2):]
            else:
                output = found[:int((len(found)+1)/2)] + ' ' + found[int((len(found)+1)/2):]
                
    return output
            
def postlen(x):
    return len(re.findall(r'\b[A-Z]{1,2}[0-9][A-Z0-9]? [0-9][ABD-HJLNP-UW-Z]{2}\b', x))
```

```python
aorig['Post'] = aorig['Address'].apply(postcode)
borig['Post'] = borig['Address'].apply(postcode)

aorig['PostLen'] = aorig['Address'].apply(postlen)
borig['PostLen'] = borig['Address'].apply(postlen)

aorig.head(2)
```

```python
print(aorig['PostLen'].value_counts())
print(borig['PostLen'].value_counts())
```

```python
def is_fc(x):
    if str(x).upper()[:2] == 'FC':
        return 'FC'
```

```python
aorig['Post2'] = aorig['CompanyNumber'].apply(is_fc)
borig['Post2'] = borig['CompanyNumber'].apply(is_fc)

aorig.head(2)
```
