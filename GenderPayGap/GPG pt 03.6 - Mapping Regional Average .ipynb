```python
import numpy as np
import pandas as pd
from pandas import pivot_table
```

```python
borig = pd.read_csv('2018clean3.csv', header=0, index_col=0, dtype={'MainSic': object})

print(borig.shape)
borig.head(2)
```

```python
post = pd.read_csv('postcodes.csv', header=0)

print(post.shape)
post.head(2)
```

```python
post['Region'].value_counts()
```

```python
post_region = post.copy()
post_region = post_region[['Postcode', 'Region']]
post_region.dropna(inplace=True)
post_region.rename(columns={"Postcode": "Post"}, inplace=True)

print(post_region.shape)
post_region.head()
```

```python
borig = pd.merge(borig, post_region, on='Post')

borig.sample(5)
```

```python
borig.tail(2)
```

```python
borig['Region'].value_counts()
```

```python
print(borig.shape[0])
print(borig['Val'].value_counts())
```

```python
map_piv = pivot_table(borig, values='DiffMeanHourlyPercent',
                     index=['Region'], aggfunc=np.mean)

map_piv.rename(columns={'DiffMeanHourlyPercent':'Diff_avg'}, inplace=True)
map_piv['Diff_avg'] = round(map_piv['Diff_avg'], 2)

map_piv
```

```python
coord = {
    'North East' : [54.949431, -1.950706],
    'North West' : [53.452868,-2.758335],
    'Yorkshire and The Humber' : [53.931954, -1.018507],
    'East Midlands' : [53.078478, -0.399826],
    'West Midlands' : [52.386524,-2.251371],
    'East of England' : [52.235806, 0.440444],
    'London' : [51.507387, -0.127768],
    'South East' : [50.998155,-0.106561],
    'South West' : [50.974859,-3.223496]
}

map_piv['lat'] = ""
map_piv['lon'] = ""

for i in map_piv.index:
    map_piv.at[i, 'lat'] = coord.get(i)[0]
    map_piv.at[i, 'lon'] = coord.get(i)[1]
    
map_piv
```

```python
map_piv.to_csv('2018regionavg.csv')
```
