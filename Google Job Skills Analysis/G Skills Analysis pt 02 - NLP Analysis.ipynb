Pre-process

```python
gged = ggms.copy() #copy to edit
```

```python
gged.head()
```

```python
gged['Responsibilities'] = gged['Responsibilities'].apply(lambda x:
                                                          " ".join(x.lower()
                                                          for x in str(x).split()))
gged['Responsibilities'].head()
```

```python
gged['Responsibilities'] = gged['Responsibilities'].str.replace('[^\w\s]', ' ')

gged['Responsibilities'].head()
```

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')

gged['Responsibilities'] = gged['Responsibilities'].apply(lambda x:
                                                          " ".join(x for x in str(x).split()
                                                          if x not in stop))
gged['Responsibilities'].head()
```

```python
freq_words = pd.Series(' '.join(gged['Responsibilities']).split()).value_counts()[:10]

freq_words
```

```python
rare_words = pd.Series(' '.join(gged['Responsibilities']).split()).value_counts()

rare_words.count()
```

```python
rare_words = rare_words[rare_words == 1]

rare_words.count()
```

Filter out rare words

```python
rare_words = list(rare_words.index)

gged['Responsibilities'] = gged['Responsibilities'].apply(lambda x:
                                                          " ".join(x for x in x.split()
                                                          if x not in rare_words))
gged['Responsibilities'].head(20)
```

Lemmatization

```python
from textblob import Word
```

```python
gged['Responsibilities'] = gged['Responsibilities'].apply(lambda x:
                                                          " ".join([Word(word).lemmatize()
                                                          for word in x.split()]))
gged['Responsibilities'].head(20)
```

```python
gged.head()
```

Pre-process Minimum Qualification column

```python
gged = gged.rename(columns={'Minimum Qualifications': 'mq'})

gged = gged.rename(columns={'Preferred Qualifications': 'pq'})

gged.head()
```

```python
gged['mq'].head()
```

```python
gged['mq'] = gged['mq'].apply(lambda x: " ".join(x.lower()
                              for x in str(x).split()))
gged['mq'].head()
```

```python
gged['mq'] = gged['mq'].str.replace('[^\w\s]', ' ')

gged['mq'].head()
```

```python
gged['mq'] = gged['mq'].apply(lambda x:
                              " ".join([Word(word).lemmatize(pos='v')
                              for word in x.split()]))
gged['mq'].head(20)
```

```python
freq_mq = pd.Series(' '.join(gged['mq']).split()).value_counts()[:18]

freq_mq
```

```python
freq_mq = list(freq_mq.index)
```

```python
gged['mq'] = gged['mq'].apply(lambda x: " ".join(x for x in x.split()
                              if x not in freq_mq))
gged['mq'].head()
```

```python
rare_mq = pd.Series(' '.join(gged['mq']).split()).value_counts()

rare_mq = rare_mq[rare_mq == 1]

rare_mq.shape
```

```python
rare_mq.head()
```

```python
gged['mq'] = gged['mq'].apply(lambda x: " ".join(x for x in x.split()
                              if x not in rare_mq))
gged['mq'].head()
```

```python
gged['mq'] = gged['mq'].apply(lambda x: " ".join(x for x in x.split()
                              if x.isdigit() == False))

gged['mq'].head()
```

```python
tf_mq = (gged['mq']).apply(lambda x: 
                          pd.value_counts(x.split(" "))
                          ).sum(axis=0).reset_index()

tf_mq.columns = ['words','tf']

tf_mq.head()
```

```python
for i,word in enumerate(tf_mq['words']):
    tf_mq.loc[i, 'idf'] = np.log(gged.shape[0]/
                        (len(gged[gged['mq'].str.contains(word)])))
tf_mq.head()
```

```python
tf_mq['tfidf'] = tf_mq['tf'] * tf_mq['idf']

tf_mq.head()
```

```python
type(tf_mq)
```

```python
mq_sorted = tf_mq.sort_values('tfidf', ascending=False)

mq_sorted.head(20)
```
