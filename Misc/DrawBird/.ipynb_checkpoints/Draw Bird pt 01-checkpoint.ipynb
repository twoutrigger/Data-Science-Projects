```python
import numpy as np
import matplotlib.pyplot as plt
im = plt.imread('draw a bird.jpeg')
plt.figure(figsize=(4.2,6))
plt.imshow(im,zorder=1)
plt.scatter(200,210, marker='$W$', s=[350], c='red', zorder=2)
plt.plot([100, 150, 225, 300, 400], [350, 600, 800, 875, 900], c='#C0C0C0')
plt.plot([100, 160, 220, 300], [300, 240, 240, 450], c='#C0C0C0')
wing = np.arange(3.0, 6.0, 0.1)
plt.plot(wing*100, np.cos(2*np.pi*wing)*150 + 600, c='pink', ls='--')
plt.show()
```
