```python
# building the training set for ML classification
```

```python
poke_fight = poke_sample[ (poke_sample['Type 1'] == 'Fighting') & (poke_sample['Stage'] == 1) ]

poke_fight.head()
```

```python
poke_electric = poke_sample[ (poke_sample['Type 1'] == 'Electric') & (poke_sample['Stage'] ==  2) ]

poke_electric.head()
```

```python
poke_poison = poke_sample[ (poke_sample['Type 1'] == 'Poison') & (poke_sample['Stage'] == 1) ]

poke_poison.info()
```

```python
poke_ml = pd.concat([poke_fight,poke_electric,poke_poison],axis=0)

poke_ml.info()
```

```python
poke_ml.drop(['Stage'],axis=1,inplace=True)

poke_ml
```

```python
sns.lmplot(x='Attack',y='Speed',data=poke_ml,hue='Type 1',fit_reg=False)
```

```python
# run ML for classification
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
poke_y = poke_ml.pop('Type 1')

poke_y
```

```python
poke_name = poke_ml.pop('Name')

poke_ml
```

```python
# function to convert Fighting / Electric / Poison to 1/2/3

def num_assign(x):
    if x == 'Fighting':
        return 0
    elif x == 'Electric':
        return 1
    elif x == 'Poison':
        return 2
```

```python
poke_y = poke_y.apply(num_assign)

poke_y
```

```python
# converted poke_y to ML needed format

poke_y = np.ravel(poke_y)

poke_y
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(poke_ml,poke_y)
```

```python
# build test x - including electric 1 / fighting 2 / poison 2

poke_test_x = poke_sample[ ((poke_sample['Type 1'] == 'Electric') & (poke_sample['Stage'] == 1)) |
                           ((poke_sample['Type 1'] == 'Fighting') & (poke_sample['Stage'] == 2)) |
                           ((poke_sample['Type 1'] == 'Poison') & (poke_sample['Stage'] == 2))
                         ]

poke_test_x
```

```python
poke_test_y = poke_test_x.pop('Type 1')

poke_test_y
```

```python
poke_test_y = poke_test_y.apply(num_assign)
```

```python
poke_test_y = np.ravel(poke_test_y)

poke_test_y
```

```python
poke_test_x.drop(['Name','Stage'],axis=1,inplace=True)

poke_test_x
```

```python
pred = logreg.score(poke_test_x,poke_test_y)
```

```python
print ('Prediction accuracy is %2.2f%%' %(pred*100))
```

```python
from sklearn.cross_validation import train_test_split
```

```python
# Fighting, Psychic, Water used for ML attempt 2

poke_sample_2 = poke_sample[ (poke_sample['Type 1'] == 'Fighting') |
                             (poke_sample['Type 1'] == 'Psychic') |
                             (poke_sample['Type 1'] == 'Water')
                           ]

poke_sample_2.info()
```

```python
sns.lmplot(x='Attack',y='Speed',data=poke_sample_2,hue='Type 1',fit_reg=False)
```

```python
poke_y_2 = poke_sample_2.pop('Type 1')

poke_y_2.head()
```

```python
# function to convert Water / Fighting / Psychic to 1/2/3

def num_assign_2(x):
    if x == 'Water':
        return 0
    elif x == 'Fighting':
        return 1
    elif x == 'Psychic':
        return 2
```

```python
poke_y_2 = poke_y_2.apply(num_assign_2)

poke_y_2 = np.ravel(poke_y_2)
```

```python
poke_sample_2.drop(['Name','Stage'],axis=1,inplace=True)

poke_sample_2.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(poke_sample_2,poke_y_2)
```

```python
logreg_2 = LogisticRegression()
```

```python
logreg_2.fit(x_train,y_train)
```

```python
pred_2 = logreg_2.score(x_test,y_test)
```

```python
print('Prediction accuracy of 2nd attempt is %2.2f%%, woot!' % (pred_2*100))
```
