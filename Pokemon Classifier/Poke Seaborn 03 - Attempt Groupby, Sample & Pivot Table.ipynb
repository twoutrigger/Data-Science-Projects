```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
poke = pd.read_csv('pokemon.csv',index_col=0,encoding='ISO-8859-1')

poke.head()
```

```python
# .sample - returns a random set of rows

poke.sample(n=5)
```

```python
# .sort_values

poke.sort_values(by='HP',ascending=False)[:5]
```

```python
poke.sort_values(by='Attack')[:5]
```

```python
# lambda for filtering

poke['Super_Attack'] = poke['Sp. Atk'].apply(lambda x: x > 100)

poke.head()
```

```python
poke[poke['Super_Attack'] == True][:5]
```

```python
# .value_counts

super_attack = poke['Super_Attack'].value_counts()

super_attack
```

```python
super_attack_portion = poke['Super_Attack'].value_counts()[1]

super_attack_portion
```

```python
All = poke['Super_Attack'].count()

All
```

```python
print('Percent of pokemon with super attacks over 100 is %2.2f%%' %
      ( (super_attack_portion/All) * 100) )
```

```python
poke['Stage'].value_counts() # see how many pokemon are in each stage
```

```python
poke.head()
```

```python
# .groupby

poke_group = poke.groupby(['Type 1','Stage']) #multi groupby

poke_group['HP'].mean()[:10]
```

```python
poke_group.size()[:10]
```

```python
poke_group_2 = poke.groupby(['Type 1','Super_Attack'])

poke_group_2.size()[:10]
```

```python
poke_group_2 = poke_group_2.size().unstack()

poke_group_2
```

```python
# plotting groupby

poke_group_2.plot(kind='barh',stacked=True)
```

```python
# pivot tables

poke_pivot = poke.pivot_table(index='Type 1',columns='Stage',
                              values='HP',aggfunc='mean')
poke_pivot.head(10)
```

```python
#not all inputs are needed; smart defaults

poke_pivot_2 = poke.pivot_table(columns='Stage')

poke_pivot_2
```

```python
poke_pivot.plot(kind='bar',stacked=True)
```
