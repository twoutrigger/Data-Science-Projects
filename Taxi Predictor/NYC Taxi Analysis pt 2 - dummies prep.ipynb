```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
txedit = pd.read_csv('edit_1_v2.csv',index_col=0)

txedit.head()
```

```python
txedit.info()
```

```python
# extract day of week from pickup datetime

from dateutil import parser
import datetime
```

```python
# weekday test

x = parser.parse('2016-03-14 17:24:55')
x.weekday()
```

```python
def day_of_week(x):
    x = parser.parse(x)
    
    return x.weekday()
```

```python
txedit['weekday'] = txedit['pickup_datetime'].apply(day_of_week)

txedit.head()
```

```python
# hour test

x = parser.parse('2016-03-14 17:24:55')
x.hour
```

```python
def hour_of_day(x):
    x = parser.parse(x)
    
    return x.hour
```

```python
txedit['pickup_hour'] = txedit['pickup_datetime'].apply(hour_of_day)

txedit.head()
```

```python
# Create dummy variable for weekdays

dumm_weekday = pd.get_dummies(txedit['weekday'],drop_first=True)

dumm_weekday.head()
```

```python
dumm_weekday.columns = ['Tues','Wed','Thu','Fri','Sat','Sun']

dumm_weekday.head()
```

```python
# Create dummy variable for hours

dumm_hour = pd.get_dummies(txedit['pickup_hour'],drop_first='True')

dumm_hour.head()
```

```python
dumm_hour.columns = ['t01','t02','t03','t04','t05','t06','t07','t08','t09','t10',
                     't11','t12','t13','t14','t15','t16','t17','t18','t19','t20',
                     't21','t22','t23']

dumm_hour.head()
```

```python
# create dataset for training

txdata = pd.concat([ txedit['passenger_count'], txedit['distance'],dumm_weekday, dumm_hour], axis=1)

txdata.head()
```

```python
# save edit_2 copy to save new dataset due to long processing time 

try:
    txdata.to_csv('edit_2.csv',mode='x')
except:
    print('File exists')
```
