```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
worig = pd.read_csv('ukcp09_gridded-land-obs-daily_timeseries_maximum-temperature_500000E_150000N_19600101-20161231.csv',
                    header=None, index_col=0)
worig.head()
```

```python
# Central London's Easting/Northing is 528409/179818
# Code below is the find the best grid (column) for London
```

```python
easting = 528409
northing = 179818
e_min = 1000000
n_min = 1000000
e_list = []
n_list = []

for i in worig.iloc[0,1:]:
    diff = abs(i-easting)
    if diff < e_min:
        e_min = diff
for i in worig.iloc[1,1:]:
    diff = abs(i-northing)
    if diff < n_min:
        n_min = diff
for col, i in enumerate(worig.iloc[0,1:]):
    if abs(i-easting) == e_min:
        e_list.append(col)
for col, i in enumerate(worig.iloc[1,1:]):
    if abs(i-northing) == n_min:
        n_list.append(col)
        
print(f'Best column fit is: {set(e_list) & set(n_list)}')
```

```python
# verify column 45 is best fit for London
worig.iloc[0:6,44]
```

```python
wedit = worig.iloc[2:,44]

wedit.rename_axis(['Date'], inplace=True)

wedit.index = pd.to_datetime(wedit.index)

print(wedit.hasnans) # check for missing values

wedit.head()
```
